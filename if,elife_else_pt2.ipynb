```python
#Atividade 01

numero=float(input("digite um numero"))
if numero >0:
  print("numero positivo")
elif numero == 0:
  print("numero nulo")
else:
  print("numero negativo")

```

```python
#Atividade 02

numero=int(input("Digite seu número"))
if numero % 2 ==0:
  print("O numero é par")
else:
  print("O numero é impar")
```
